### increase thickness with image dilation

```python
kernal = np.ones((5,5), np.uint8)
```

```python
imgDial = cv2.dilate(imgCanny, kernal, iterations=1)
```

```python
cv2.imshow("APJ [Window_Name]", imgDial)
cv2.waitKey(2000)
destroy()
```

```python
plt.imshow(imgDial)
```

```python
imgDial1 = cv2.dilate(imgCanny, kernal, iterations=3)
cv2.imshow("APJ [Window_Name]", imgDial1)
cv2.waitKey(2000)
destroy()
plt.imshow(imgDial1)
```

## Erode = oppposite of dilation

```python
img_erode = cv2.erode(imgDial, kernal, iterations=1)
cv2.imshow("APJ [Window_Name]", img_erode)
cv2.waitKey(2000)
destroy()
plt.imshow(img_erode)
```

# WARP PERSPECTIVE


```python
img_joker = cv2.imread("./data/joker.jpg")
```

```python
plt.imshow(img_joker)
```

```python
width, height = 250, 350
points = np.float32([[250, 187], [210, 500], [1000, 450], [650, 870]])
points_1 = np.float32([[0, 0], [width, 0], [0, height], [width, height]])
```

```python
matrix = cv2.getPerspectiveTransform(points, points_1)
```

```python
imgOutput = cv2.warpPerspective(img_joker, matrix, (width, height))
```

```python
#cv2.imshow('output', imgOutput)
#cv2.waitKey(2000)
#destroy()
plt.imshow(imgOutput)
```
